```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv('/Users/admin/flatiron/ds_mod1/ds-challenge-05_seattle-ds-career-040119/data.csv',header=None)
df.columns = df.columns = ['day_1', 'day_2', 'day_3', 'day_4', 'day_5', 'day_6']
```

```python
print(df)
```

```python
df['day_1'][7] = '0:00 AM'
df
```

```python
df = df.applymap(lambda x : pd.to_datetime(x).strftime('%H:%M:%S'))
df
```

```python
df.info()
```

```python
day_1 = list(df.day_1)
day_2 = list(df.day_2)
day_3 = list(df.day_3)
day_4 = list(df.day_4)
day_5 = list(df.day_5)
day_6 = list(df.day_6)
```

```python
x = day_1 + day_2 + day_3 + day_4 + day_5 + day_6
```

```python

plt.figure(figsize=(35,10))
plt.hist(x, bins = 20)
plt.xlabel('Times', fontsize=35)
plt.ylabel('Frequency', fontsize=30)
plt.title('Arrival Times', fontsize=40)
```
